```python
import pandas as pd

a = pd.read_csv('dataset_TIST2015_Checkins.txt', sep='\t', header=None, names=['user', 'locid', 'time', 'offset', 'non'])
b = pd.read_csv('dataset_TIST2015_POIs.txt', sep='\t', header=None, names=['locid', 'lat', 'lon', 'name', 'category', 'country', 'city', 'popular'])
c = a.join(b.set_index('locid'), on='locid')

#filter the dataset c to only regions in UCLA
latmin, lonmin,  latmax,lonmax = 34.067089, -118.445003, 34.075976, -118.439485
c = c[(c['lat'] > latmin) & (c['lat'] < latmax) & (c['lon'] > lonmin) & (c['lon'] < lonmax)]
c["time"] = pd.to_datetime(c['time'].str[:20]+c['time'].str[26:])
```

```python
c.loc[:,['user','time','lat','lon','name']].to_csv('ucla_checkins.csv', index=False, header=False)
```
